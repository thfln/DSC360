## Ingesting Unstructured Text

**1. Read the text from the provided PDF file using Python. Print the results.**

```python
# Define the path where the PDF file is located
pdf = r"C:\Users\Amanda Heflin\OneDrive\Documents\BellevueUniversity\Week_3_No_Tables (1).pdf"

# Define a function to extract the text
def extract_text(pdf):
    # Open in binary read mode
    with open(pdf, 'rb') as file:
        # Utilize imported PDF reader
        reader = PyPDF2.PdfReader(file)

        # Extract text
        pdf_text = ""
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            pdf_text += page.extract_text()

    return pdf_text

# Extract and print the results
pdf_content = extract_text(pdf)
print(pdf_content)
```

**4. Assuming you installed Tesseract, use pytesseract to read the Bowers text image provided. Then, format the resulting text by removing the extraneous characters. Finally, use spaCy to print out the tokens from that text (the text, part of speech, and dependency). (HINT: you may need to print the tokens to see what characters you need to remove.)**

```python
# Initialize tesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\Users\Amanda Heflin\OneDrive\Desktop\theflin\tesseract.exe'
# Load spaCy model
nlp = spacy.load('en_core_web_sm')

# Define function to fetch image from URL and convert to PIL
def image_from_file(image_path):
    img = Image.open(image_path)

    # Perform OCT on image
    raw_text = pytesseract.image_to_string(img)

    # Remove non-alphanumeric characters (excluding spaces)
    clean_text = ''.join(char for char in raw_text if char.isalnum() or char.isspace())
    return clean_text

# Define function to analyze text with spaCy
def analyze_text(text):
    doc = nlp(text)

    # Print tokens to ensure validity
    for token in doc:
        print(f'Token: {token.text}, POS: {token.pos_}, Dependency: {token.dep_}')

# Define the image path
image_file_path = r'C:\Users\Amanda Heflin\OneDrive\Documents\BellevueUniversity\bowers.jpg'

# Get text from image
text_from_image = image_from_file(image_file_path)

# Analyze original and cleaned text
print("Original text from image:")
print(text_from_image)

print("\nAnalyzed text with spaCy:")
analyze_text(text_from_image)
```
